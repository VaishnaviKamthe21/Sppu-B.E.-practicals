```python
import math
def haversine(lat1,lon1,lat2,lon2):
    distance = []
    for pos in range(len(lon1)):
        long1,lati1,long2,lati2 = map(math.radians,[lon1[pos],lat1[pos],lon2[pos],lat2[pos]])
        dist_long = long2 - long1
        dist_lati = lati2 - lati1
        a = math.sin(dist_lati/2)**2 + math.cos(lati1) * math.cos(lati2) * math.sin(dist_long/2)**2
        c = 2 * math.asin(math.sqrt(a))*6371
        distance.append(c)
    return distance

df['distance'] = haversine(df['pickup_latitude'].to_numpy(), df['pickup_longitude'].to_numpy(), df['dropoff_latitude'].to_numpy(), df['dropoff_longitude'].to_numpy())
```

```python
df = df.assign(pickup_hr = df.pickup_datetime.dt.hour,
               day= df.pickup_datetime.dt.day,
               month = df.pickup_datetime.dt.month,
               year = df.pickup_datetime.dt.year,
               day_of_week = df.pickup_datetime.dt.dayofweek)
df.drop(df[['pickup_datetime']], axis=1, inplace=True)
```

```python
Q1 = df[ 'fare_amount'].quantile(0.25)
Q3 = df[ 'fare_amount'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

lower_arr = np.where(df['fare_amount']<=lower)[0]
upper_arr = np.where(df['fare_amount']>=upper)[0]

df.drop(index=upper_arr, inplace=True)
df.drop(index=lower_arr, inplace=True)

df.shape
```

```python
df = df.reset_index()
del df['index']
```

```python
Q1 = df[ 'distance'].quantile(0.25)
Q3 = df[ 'distance'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

lower_arr = np.where(df['distance']<=lower)[0]
upper_arr = np.where(df['distance']>=upper)[0]

df.drop(index=upper_arr, inplace=True)
df.drop(index=lower_arr, inplace=True)

df.shape
```

```python
x = df[['distance']]
y = df['fare_amount']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
x_train, x_test , y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=48)
```

```python
model1 = LinearRegression()
model1.fit(x_train,y_train)
```

```python
y_pred = model1.predict(x_test)
```

```python
sns.regplot(x = y_test, y = y_pred)
plt.show()
```

```python
print(f"MSE = {metrics.mean_squared_error(y_test,y_pred)}")
print(f"MAE = {metrics.mean_absolute_error(y_test,y_pred)}")
print(f"RMSE = {metrics.root_mean_squared_error(y_test,y_pred)}")
print(f"R2 = {metrics.r2_score(y_test,y_pred)}")
```

```python
model2 = RandomForestRegressor()
model2.fit(x_train,y_train)
```

```python
y_pred = model2.predict(x_test)
```

```python
sns.regplot(x = y_test, y = y_pred)
plt.show()
```

```python
print(f"MSE = {metrics.mean_squared_error(y_test,y_pred)}")
print(f"MAE = {metrics.mean_absolute_error(y_test,y_pred)}")
print(f"RMSE = {metrics.root_mean_squared_error(y_test,y_pred)}")
print(f"R2 = {metrics.r2_score(y_test,y_pred)}")
```
